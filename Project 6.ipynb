# Attacks by date over time

```python
import project6 as ld

import pandas as pd
import matplotlib.pyplot as plt

conn = ld.connect_db('db.sqlite')
data = pd.read_sql_query("SELECT * from dates ORDER BY attack_count asc",conn)

ax = plt.gca()

data.plot(kind='line',x='event_timestamp',y='attack_count',ax=ax)
data.head()
plt.show()
```
